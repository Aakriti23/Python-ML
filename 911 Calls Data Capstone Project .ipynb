## Creating new features

** In the titles column there are "Reasons/Departments" specified before the title code. These are EMS, Fire, and Traffic. Using .apply() with a custom lambda expression to create a new column called "Reason" that contains this string value.** 

**For example, if the title column value is EMS: BACK PAINS/INJURY , the Reason column value would be EMS. **

```python
df['Reason']=df['title'].apply(lambda x: x.split(':')[0])
```

```python
df
```

** What is the most common Reason for a 911 call based off of this new column? **

```python
df['Reason'].value_counts()
```

```python
sns.countplot(x='Reason',data=df)
```

___
** Checking the data type of the objects in the timeStamp column **

```python
type(df['timeStamp'].iloc[0])
```

** Since the data type of the objects in the timeStamp column is string, we use [pd.to_datetime](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.to_datetime.html) to convert the column from strings to DateTime objects. **

```python
df['timeStamp']=pd.to_datetime(df['timeStamp'])
```

```python
type(df['timeStamp'].iloc[0])
```


** Now that the timestamp column are actually DateTime objects, I create three new columns titled "Hour", "Month" & "Day of the week" using .apply() & functions that come for datetime objects. **

```python


df['hour']=df['timeStamp'].apply(lambda time:time.hour)
df['month']=df['timeStamp'].apply(lambda time:time.month)
df['dayofweek']=df['timeStamp'].apply(lambda time:time.dayofweek)
```

```python
df
```

** Using the .map() with this dictionary to map the actual string names to the day of the week: **

    dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
```

```python
df['dayofweek']=df['dayofweek'].map(dmap)
```

```python
df
```

** Using seaborn to create a countplot of the Day of Week column with the hue based off of the Reason column. **

** Gives the count of the number of calls made on each weekday separated by the "Reasons" column. **

```python
sns.countplot(x='dayofweek',hue='Reason', data=df)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

**Making the same countplot for Month:**

** It shows that most calls are made in January and they decrease considerably from January to December. **

```python
sns.countplot(x='month',hue='Reason', data=df)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

**There is something strange about the count plot. Some of the months are missing.**

_____

** I tried to fill in this information by plotting the information in another way, possibly a simple line plot that fills in the missing months. In order to do this, I created a groupby object. **

** Creating a gropuby object called dg, where I grouped the DataFrame by the month column and used the count() method for aggregation. **

```python
dg=df.groupby(by='month').count()
dg
```

** Plotting off of the dataframe indicating the count of calls per month. **

** The relationship between the month and number of calls is evident from the line graph. **

```python

dg['twp'].plot()
```

** Using seaborn's lmplot() to create a linear fit on the number of calls per month. **

```python
dm=dg.reset_index()
```

```python
sns.lmplot(x='month',y='twp', data=dm)
```

**Creating a new column called 'Date' that contains the date from the timeStamp column. **

```python
df['date']=df['timeStamp'].apply(lambda time:time.date())
```

** Grouping this Date column with the count() aggregate and creating a plot of counts of 911 calls wrt to dates.**

```python
df.groupby(by='date').count()['twp'].plot(color='blue')
plt.tight_layout()
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

**  Creating 3 separate plots with each plot representing a Reason for the 911 call**

```python
df[df['Reason']=='Traffic'].groupby('date').count()['twp'].plot()
plt.title('Traffic')
plt.tight_layout()
```

```python
df[df['Reason']=='Traffic'].groupby('Date').count()['twp'].plot()
plt.title('Traffic')
plt.tight_layout()
```

```python
df[df['Reason']=='Fire'].groupby('date').count()['twp'].plot()
plt.title('Traffic')
plt.tight_layout()
```

```python
df[df['Reason']=='EMS'].groupby('date').count()['twp'].plot()
plt.title('Traffic')
plt.tight_layout()
```

## Creating Heatmaps

____
** Restructuring the dataframe so that the columns become the Hours and the Index becomes the Day of the Week. Using groupby with an [unstack](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.unstack.html) method to do the same.**

```python
dayHour = df.groupby(by=['dayofweek','hour']).count()['Reason'].unstack()
dayHour.head()
```

** Creating a HeatMap using this new DataFrame. **

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayHour,cmap='viridis')
```

** Creating a clustermap using this DataFrame. **

```python
plt.figure(figsize=(12,6))
sns.clustermap(dayHour,cmap='viridis')
```

** Repeating the above plots and operations, for a DataFrame that shows the Month as the column. **

```python
dayMonth = df.groupby(by=['dayofweek','month']).count()['Reason'].unstack()
dayMonth.head()
```

```python
plt.figure(figsize=(12,6))
sns.clustermap(dayMonth,cmap='viridis')
```
