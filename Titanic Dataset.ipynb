# Titanic Dataset

Will be focussing on visualisation using seaborn.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
#Creating a joint plot to study the relationship between the fare and age of the people travelling in the ship.

sns.jointplot(x="fare",y="age", data=titanic)
```

```python
#Creating a distplot to study the fares. It is clear that the fare is between 0-100 for majority of the population.

sns.distplot(titanic['fare'],kde=False, bins=35, color='red')
```

```python
# Creating a boxplot 

sns.boxplot(x='class',y='age', data=titanic,palette='rainbow')
```

```python
# Creating a swarmplot

sns.swarmplot(x='class',y='age', data=titanic,palette='Set2')
```

```python
# Creating a countplot. It is evident that there were more number of males than there were females.

sns.countplot(x='sex', data=titanic)
```

```python
# Creating a heatmap to study the correlation between all the variables of the data.

ta=titanic.corr()
sns.heatmap(ta, cmap='coolwarm')
```

```python
# Creating a FacetGrid.

g=sns.FacetGrid(data=titanic, col='sex',margin_titles=True)
g.map(plt.hist, 'age', bins=10,edgecolor='Black')
```
